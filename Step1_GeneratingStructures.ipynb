```python
def StructureGenerator(upperbound,dbname):
    from clease import Concentration
    from clease import CEBulk
    from clease import NewStructures
    from ase.calculators.emt import EMT
    from ase.db import connect
    from clease.tools import update_db
    import numpy as np
    from clease import Evaluate
    
    #Define Concentrations and Setting
    #we set our conditions
    db_name=dbname
    #Check if DB exists
    db=connect('/home/s376r951/Latest_Project/VASPDFT/StartingFresh/%s' %db_name)
    if len(db)==0:
        print('DB does not exist, continuing..')
        skipper=0
    else:
        print('DB exists already, skipping initial generation')
        skipper=1
    ##Skipper tells us if DB exists, 1 for true 0 for false. 
    #Now we define our settings and concentration range
    conc=Concentration(basis_elements=[['Au', 'Cu']])
    conc.set_conc_formula_unit(formulas=["Au<x>Cu<1-x>"], variable_range={"x": (0, 1.0)})
    settings=CEBulk(crystalstructure='fcc', a=3.8, size=[3,3,3], 
                    concentration=conc, db_name=db_name,
                    max_cluster_size=4, 
                    max_cluster_dia=[6.0,5.0,5.0])
    cv_list=[] 
    cv=0
    if skipper==0:
        #generates initial structures if db does not exist
        ns=NewStructures(settings, generation_number=1, struct_per_gen=10)
        ns.generate_initial_pool()
    else:
        print('skipper triggered,  skipping initial generation')
    
    
    #Now we generate some random structures
    db=connect('/home/s376r951/Latest_Project/VASPDFT/StartingFresh/%s' %db_name)
    for i in range(1,upperbound):
        gen_num=i
        rs=NewStructures(settings, generation_number=gen_num+1, struct_per_gen=1)
        rs.generate_initial_pool()
    print('generation of random structures is complete')
    print('number of generated structures is %s'%len(db))
        
    
```

```python
StructureGenerator(40, "FixedCell.db")
```
